```python
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import HTML
```

```python
path='../QiuShi_Spider/qiushi_user_info.json'
data = [json.loads(line) for line in open(path)]
data = pd.DataFrame(data)
```

```python
#del data['followers']
#data
```

```python
data.rename(columns=({'username':u'用户名',u'关注数:':u'关注数',u'婚姻:':u'婚姻',u'故乡:':u'故乡',u'星座:':u'星座',u'笑脸:':u'笑脸',u'粉丝数:':u'粉丝数',u'糗事:':u'糗事',u'糗事精选:':u'糗事精选',u'糗龄:':u'糗龄',u'职业:':u'职业',u'评论:':u'评论'}),inplace=True)
```

```python
data.replace(np.nan,-1,inplace=True)
```

```python
data['id'] = data['id'].astype(int)
data[u'笑脸'] = data[u'笑脸'].astype(int)
data[u'关注数'] = data[u'关注数'].astype(int)
data[u'粉丝数'] = data[u'粉丝数'].astype(int)
data[u'糗事'] = data[u'糗事'].astype(int)
data[u'糗事精选'] = data[u'糗事精选'].astype(int)
data[u'评论'] = data[u'评论'].astype(int)
```

```python
data.drop_duplicates(subset='id',inplace=True)
data.sort_values(by=u'关注数',ascending=False)
```

```python
data.sort_values(by=u'粉丝数',ascending=False)
```

```python
def marriage(df,column=u'婚姻'):
    return df.sort_values(column, ascending=False)[:][u'婚姻'].count()
marriage_col = data[u'婚姻']
marriage_col = pd.DataFrame(marriage_col)
marriage_col.replace(-1,'unknown',inplace=True)
#marriage_col.replace('secret',u'未知',inplace=True)
#marriage_col.replace('single',u'单身',inplace=True)
#marriage_col.replace('married',u'已婚',inplace=True)
#marriage_col.replace('inlove',u'热恋',inplace=True)
marriage_col_data = marriage_col.groupby(by=u'婚姻').apply(marriage)
marriage_col_data
```

```python
def hometown(df,column=u'故乡'):
    return df.sort_values(column, ascending=False)[:][u'故乡'].count()
data.groupby(by=u'故乡').apply(hometown)
```

```python
col = data[[u'粉丝数',u'关注数']]
col = pd.DataFrame(col)
col.rename(columns=({u'粉丝数':'fans',u'关注数':'follower'}),inplace=True)
col = col[col['fans']<20000]
col = col[col['fans']>-1]
col = col[col['follower']<4000]
col.sort_values(by='fans',ascending=False)
col.drop(col['fans']>2,inplace=True)
col.plot(x='fans', y='follower', kind='scatter')
```

```python
labels1=['inlove','married','secret','single','unknown']
marriage_col_data
plt.axes(aspect=1)
plt.pie(x=marriage_col_data.values,labels=labels1,autopct='%.0f%%',explode=None,shadow=True)
plt.show()
```
